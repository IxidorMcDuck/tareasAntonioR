Haz un gas ideal:

    Partículas en 2D (x y v que son vectores con 2 entradas)
    Hacer una función que genere al azar una partícula
    Hacer una función que agregue una partícula al Gas (cuidar que el número de partículas se actualice)
    Hacer una función que genere un gas
    Mover el gas: actualizar todas las partículas y hacer algo en las fronteras (de preferencia, colisiones elásticas con vectores)
    [Para empezar: colisiones inelásticas: si se sale, se queda donde está; y/o condiciones periódicas: si se sale, se regresa del otro lado]



```julia
type Particula
    posx
    posy
    velx
    vely
    mass
end

type Gas
    N::Int
    moleculas::Vector{Particula}
end

function mover!(part,dt)
    part.posx=part.posx+part.velx*dt
    part.posy=part.posy+part.vely*dt
    part
end

function Gas(N,m)
    G=Array(Particula,N)
    for i=1:N
        G[i]=Particula(rand(),rand(),((-1)^(rand(0:1)))*rand(),((-1)^(rand(0:1)))*rand(),m)
        #G[i]=Particula(BigFloat(rand()),BigFloat(rand()),BigFloat(rand()),BigFloat(rand()),m)
    end
    Gas(N,G)
end

function moveGas!(Ens,dt)
    for i=1:Ens.N
        mover!(Ens.moleculas[i],dt)
    end
    Ens
end

function addParticle!(Ens)
    m=Ens.moleculas[1].mass
    Ens.N=Ens.N+1
    push!(Ens.moleculas,Particula(rand(),rand(),((-1)^(rand(0:1)))*rand(),((-1)^(rand(0:1)))*rand(),m))
    Ens
end

function EvolucionPeriodica!(Ens,steps)
    for i=1:steps
           moveGas!(Ens,2.0^(-5))
        for j=1:Ens.N
            if Ens.moleculas[j].posx >= 1
                Ens.moleculas[j].posx=Ens.moleculas[j].posx-1
            end
            if Ens.moleculas[j].posx <= 0
                Ens.moleculas[j].posx=Ens.moleculas[j].posx+1
            end
            
            if Ens.moleculas[j].posy >= 1
                Ens.moleculas[j].posy=Ens.moleculas[j].posy-1
            end
            if Ens.moleculas[j].posy <= 0
                Ens.moleculas[j].posy=Ens.moleculas[j].posy+1
            end
        end
    println(Ens)
    end
     
    end

function EvolucionElastica!(Ens,steps)
    for i=1:steps
           moveGas!(Ens,2.0^(-5))
        for j=1:Ens.N
            if Ens.moleculas[j].posx >= 1
                Ens.moleculas[j].velx=-Ens.moleculas[j].velx
            end
            if Ens.moleculas[j].posx <= 0
                Ens.moleculas[j].velx=-Ens.moleculas[j].velx
            end
            
            if Ens.moleculas[j].posy >= 1
                Ens.moleculas[j].vely=-Ens.moleculas[j].vely
            end
            if Ens.moleculas[j].posy <= 0
                Ens.moleculas[j].vely=-Ens.moleculas[j].vely
            end
        end
   println(Ens)
    end
    
    end

function EvolucionEstatica!(Ens,steps)
    for i=1:steps
           for j=1:Ens.N
            if Ens.moleculas[j].posx <= 1 && Ens.moleculas[j].posx >=0 && Ens.moleculas[j].posy <= 1 && Ens.moleculas[j].posy >=0
                mover!(Ens.moleculas[j],2.0^(-5))
            end
            
            end
   println(Ens)
    end
    
    end
```

```julia
EvolucionElastica!(z,100)
```
